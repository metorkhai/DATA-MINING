```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline
```

```python
print(sns.get_dataset_names())
```

```python
tip_df = sns.load_dataset('tips')
```

```python
tip_df.head()
```

```python
print(tip_df.shape)
```

```python
sns.set_theme( )
```

```python
sns.histplot(tip_df['total_bill'])
```

```python
tip_df["total_bill"].value_counts().sort_values(ascending=False)
```

```python
sns.kdeplot(tip_df['total_bill'])
```

```python
sns.displot(data=tip_df, x="total_bill", col="time",  kde=True)
```

```python
sns.barplot(data=tip_df, x="sex", y="tip", estimator=np.mean)
```

```python

tip_df['sex'].value_counts()
```

```python
sns.countplot(data=tip_df, x="sex")
```

```python
sns.boxplot(data=tip_df, x="day", y="total_bill", hue="sex", palette="Blues")
plt.legend(loc=0)
```

```python
new_fg = sns.FacetGrid(data=tip_df, col="time", row="smoker")
new_fg.map(sns.scatterplot, 'total_bill', 'tip')
```

```python
kws=dict(s=50, edgecolor="y")
new_tip = sns.FacetGrid(data=tip_df, col="sex", hue="smoker", col_order=["Female", "Male"],
                        palette="Set1",
                        height=4, aspect=1.5)
new_tip.map(sns.scatterplot, 'total_bill', 'tip',**kws)
new_tip.add_legend()
```
