```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cercreal_df = pd.read_csv("/content/sample_data/cereal.csv")
```

```python
cercreal_df.head()
```

```python
fields = ['shelf', 'weight', 'cups', 'rating']
recreal_df_new = cercreal_df.drop(fields, axis=1)
recreal_df_new.head()
```

```python
recreal_corr = recreal_df_new.corr()
recreal_corr
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(recreal_corr, annot=True)
plt.tight_layout()
```
